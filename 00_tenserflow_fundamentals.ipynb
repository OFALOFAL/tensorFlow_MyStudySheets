```python
tensor_1 = tf.constant([[3, 7],
                        [1, 3],
                        [2, 8]])
tensor_2 = tf.constant([[3, 7],
                        [1, 3],
                        [2, 8]])
# won't work becouse of the shape
# tensor_1 @ tensor_2

# to make matrix multip work the shape of matrix must be flipped in one of the matrix

# slow methods:
# tensor_2 = tf.constant([[3, 7, 1], [2, 8, 3]])
# or
# tensor_2 = tf.reshape(tensor_2, shape=(2, 3))
# tensor_1 @ tensor_2

# normal method:
print(tensor_1 @ tf.reshape(tensor_2, shape=(2, 3)))

# or
print(tf.transpose(tensor_1) @ tensor_2)
#note: the shape of this multipycation is different becouse shape of the right-site tensor is directed by outer sizes of both left-side tensors
# shape is different not becouse of transpose but becouse we do transpose on 1st tensor

# note 2: transpose and reshape are different in that reshape is random and transpose is not
```

```python
# min, max, mean, sum
tf.reduce_min(tensor_1), tf.reduce_max(tensor_1), tf.reduce_mean(tensor_1), tf.reduce_sum(tensor_1)
```

```python
# standart deviation and variance using math
tf.math.reduce_std(tf.cast(tensor_1, dtype=tf.float32))
tf.math.reduce_variance(tensor_1)
```

```python
# variance and normal distribution using tfp
import tensorflow_probability as tfp
tfp.stats.variance(tensor_1)
tfp.distributions.Normal(tensor_1, scale=tensor_2)
```
